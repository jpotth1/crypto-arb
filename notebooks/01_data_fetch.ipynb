```python
import ccxt
import pandas as pd
from src.spread import compute_spread

# Initialize exchanges
binance = ccxt.binanceus({
    'enableRateLimit': True
})
kraken  = ccxt.kraken()

# Fetch 500 one-minute candles
df_bin = pd.DataFrame(
    binance.fetch_ohlcv('BTC/USD', '1m', limit=500),
    columns=['ts','open','high','low','close','vol']
)
df_kra = pd.DataFrame(
    kraken.fetch_ohlcv('BTC/USD',   '1m', limit=500),
    columns=['ts','open','high','low','close','vol']
)

df_bin.to_csv('../data/binance_1m.csv', index=False)
df_kra.to_csv('../data/kraken_1m.csv',  index=False)

# Convert timestamps to datetime
df_bin['ts'] = pd.to_datetime(df_bin['ts'], unit='ms')
df_kra['ts'] = pd.to_datetime(df_kra['ts'], unit='ms')

# Inspect
df_bin.head(), df_kra.head()

# Compute spreads
df = compute_spread('../data/binance_1m.csv','../data/kraken_1m.csv')
df[['ts','close_bin','close_kra','spread','signal']].head(10)
```
