```python
import pandas as pd
import matplotlib.pyplot as plt
from src.spread import compute_spread

# Load the merged DataFrame with spreads and signals
df = compute_spread('../data/binance_1m.csv', '../data/kraken_1m.csv')

# Parameters
trade_size = 1.0  # units of asset
taker_fee_rate = 0.001  # 0.1% fee per trade side

# Simulate P&L for each signal
df['pnl'] = 0.0
trade_indices = df['signal']
df.loc[trade_indices, 'pnl'] = (
    (df.loc[trade_indices, 'close_bin'] - df.loc[trade_indices, 'close_kra']) * trade_size
    - trade_size * taker_fee_rate * (df.loc[trade_indices, 'close_bin'] + df.loc[trade_indices, 'close_kra'])
)

# Cumulative P&L
df['cum_pnl'] = df['pnl'].cumsum()

# Display the first few results
from IPython.display import display
display(df[['ts','close_bin','close_kra','spread','signal','pnl']].head(10))

# Plot cumulative P&L
plt.figure(figsize=(10, 5))
plt.plot(df['ts'], df['cum_pnl'])
plt.title("Cumulative P&L of Simple Arb Strategy (Paper Trades)")
plt.xlabel("Timestamp")
plt.ylabel("Cumulative P&L (USD)")
plt.tight_layout()
plt.show()
```
